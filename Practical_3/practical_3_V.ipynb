```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('teachers_rating_data (4).csv')
df.head()
```

# Q1 Identify all duplicate cases using prof. Using all observations, find the average and standard deviation for age. Repeat the analysis by first filtering the data set to include one observation for each instructor with a total number of observations restricted to 94.

```python
duplicates = df[df.duplicated(subset='prof', keep=False)]
print("Duplicate cases based on 'prof':")
print(duplicates)

mean_age_all = df['Age'].mean()
std_age_all = df['Age'].std()
print(f"All observations - Mean Age: {mean_age_all:.2f}, Std Dev: {std_age_all:.2f}")

df_unique = df.drop_duplicates(subset='prof', keep='first').head(94)
mean_age_unique = df_unique['Age'].mean()
std_age_unique = df_unique['Age'].std()
print(f"Unique professors (94) - Mean Age: {mean_age_unique:.2f}, Std Dev: {std_age_unique:.2f}")
```

# Q2 Using a bar chart, demonstrate if instructors teaching lower-division courses receive higher average teaching evaluations.

```python
import matplotlib.pyplot as plt

avg_eval_by_course = df.groupby('CourseLevel')['Evaluation'].mean()
avg_eval_by_course.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average Teaching Evaluation by Course Level')
plt.ylabel('Average Evaluation')
plt.xlabel('Course Level')
plt.show()
```

# Q3 Plot the relationship between age and teaching evaluation scores.

```python
plt.scatter(df['Age'], df['Evaluation'], color='green', alpha=0.6)
plt.title('Age vs Teaching Evaluation')
plt.xlabel('Age')
plt.ylabel('Evaluation Score')
plt.grid(True)
plt.show()
```

# Q4 Using gender-differentiated scatter plots, plot the relationship between age and teaching evaluation scores.

```python
import seaborn as sns

sns.scatterplot(data=df, x='Age', y='Evaluation', hue='Gender', style='Gender', s=80)
plt.title('Age vs Evaluation by Gender')
plt.show()
```

# Q5 Create a scatterplot of age and evaluation scores, differentiated by gender and tenure.

```python
sns.scatterplot(data=df, x='Age', y='Evaluation', hue='Gender', style='Tenure', s=80)
plt.title('Age vs Evaluation by Gender and Tenure')
plt.show()
```
